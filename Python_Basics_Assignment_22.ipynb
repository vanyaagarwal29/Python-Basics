Let's go through each code snippet one by one:

1. Code snippet:
```python
X = 'iNeuron'
def func():
    print(X)

func()
```

Output:
```
iNeuron
```

Explanation:
- The variable `X` is defined outside the function `func`, making it a global variable.
- Inside the function `func`, we print the value of `X`, which is `'iNeuron'`.
- When we call the function `func()`, it prints the value of the global variable `X`, which is `'iNeuron'`.

2. Code snippet:
```python
X = 'iNeuron'
def func():
    X = 'NI!'

func()
print(X)
```

Output:
```
iNeuron
```

Explanation:
- The variable `X` is defined outside the function `func`, making it a global variable.
- Inside the function `func`, we define a local variable `X` with the value `'NI!'`.
- When we call the function `func()`, it defines a local variable `X` with the value `'NI!'`, but it does not affect the global variable `X`.
- After the function call, when we print the value of the global variable `X`, it remains unchanged and prints `'iNeuron'`.

4. Code snippet:
```python
X = 'iNeuron'
def func():
    global X
    X = 'NI'

func()
print(X)
```

Output:
```
NI
```

Explanation:
- The variable `X` is defined outside the function `func`, making it a global variable.
- Inside the function `func`, we use the `global` keyword to indicate that we want to modify the global variable `X`.
- When we call the function `func()`, it changes the value of the global variable `X` to `'NI'`.
- After the function call, when we print the value of the global variable `X`, it now reflects the modified value and prints `'NI'`.

5. Code snippet:
```python
X = 'iNeuron'
def func():
    X = 'NI'
    def nested():
        print(X)
    nested()

func()
print(X)
```

Output:
```
NI
iNeuron
```

Explanation:
- The variable `X` is defined outside the function `func`, making it a global variable.
- Inside the function `func`, we define a local variable `X` with the value `'NI'`.
- Inside `func`, we define another function `nested` that prints the value of the local variable `X`.
- When we call the function `func()`, it calls the nested function, and the nested function prints the value of the local variable `X`, which is `'NI'`.
- After the function calls are finished, when we print the value of the global variable `X`, it remains unchanged and prints `'iNeuron'`.

6. Code snippet:
```python
def func():
    X = 'NI'
    def nested():
        nonlocal X
        X = 'Spam'
    nested()
    print(X)

func()
```

Output (Python 3):
```
Spam
```

Explanation:
- The function `func` defines a local variable `X` with the value `'NI'`.
- Inside `func`, we define another function `nested`, and we use the `nonlocal` keyword to indicate that we want to modify the `X` from the enclosing (outer) scope (in this case, the `func` scope).
- Inside the nested function, we change the value of the nonlocal variable `X` to `'Spam'`.
- When we call the function `func()`, it calls the nested function, and the nested function modifies the nonlocal variable `X`.
- After the function call, when we print the value of the local variable `X` in the `func` scope, it now reflects the modified value and prints `'Spam'`.
